```python
import pandas as pd
import matplotlib.pyplot as plt

# read the horror dataset from a URL
url = "https://raw.githubusercontent.com/merkington/fffff/main/horror.csv"
horror = pd.read_csv(url)

# extract the year, rating, and votes columns
year = horror['year']
rating = horror['rating']
votes = horror['votes']

# calculate the average of votes/rating for each year
avg_votes_per_rating = horror.groupby('year')['votes'].sum() / horror.groupby('year')['rating'].sum()

# create a horizontal bar graph to show the result
plt.figure(figsize=(16, 24))
plt.barh(avg_votes_per_rating.index, avg_votes_per_rating.values)
plt.xlabel('Average votes/rating')
plt.ylabel('Year')
plt.title('How the average of votes/rating changes over time')
plt.show()
```
