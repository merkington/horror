```python
import pandas as pd
import matplotlib.pyplot as plt

# read the horror dataset from a URL
url = "https://raw.githubusercontent.com/merkington/fffff/main/horror.csv"
horror = pd.read_csv(url)

# extract the year, rating, and votes columns
year = horror['year']
rating = horror['rating']
votes = horror['votes']

# calculate the average of votes/rating for each year
avr = horror.groupby('year')['votes'].sum() / horror.groupby('year')['rating'].sum()

cat = horror['genre'].value_counts()
catcut = cat.quantile(0.00001)
cat = cat[cat >= catcut]
plt.subplot(1, 2, 1)
plt.pie(cat, labels=cat.index)

Year = horror['year'].value_counts()
cutoff = Year.quantile(0.9)
Year = Year[Year >= cutoff]

#genl = horror['genre']
plt.subplot(1, 2, 2)
plt.pie(Year, labels=Year.index)
plt.show()

# create a horizontal bar graph to show the result
plt.figure(figsize=(16, 24)) 
plt.barh(avr.index, avr.values)
plt.xlabel('Average votes/rating')
plt.ylabel('Year')
plt.title('How the average of votes/rating changes over time')
#plt.show()
```
